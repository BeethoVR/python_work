# Gráficas de línea
<p style='text-align: justify;'>Las <b><font color = green>gráficas de línea</font></b> son una representación visual de valores que se modifican sobre el intervalo de una variable continua usualmente en los problemas del mundo real se utiliza el tiempo como dicha variable.</p> 

<p style='text-align: justify;'>Las gráficas de línea generalmente se utilizan para representar tendencias del comportamiento de una variable o para evidenciar la relación de dos o más variables durante el mismo intervalo.</p>
<br>
<center><img src="img/Linea1.png" width = "70%"></center>
<br>
<p style='text-align: justify;'>En nuestro primer ejemplo hemos representado todo el año de ventas de una compañía. Los elementos que componen nuestro primer gráfico de línea son:</p> 
<p style='text-align: justify;'>1.	El <font color = green><b>eje horizontal<b></font> que representa el intervalo de tiempo en el cuál se analizará el cambio en la variable que buscamos estudiar.</p>
<p style='text-align: justify;'>2.	El <font color = green><b>eje vertical<b></font> que representa el intervalo continuo de ventas que se realizaron durante el periodo de tiempo analizado. Generalmente este eje se delimita utilizando el valor máximo y el valor mínimo que se la variable analizada puede obtener.</p>
<p style='text-align: justify;'>3.	Los <font color = green><b>puntos de datos<b></font> en la cuadrícula cartesiana, donde cada coordenada de un punto representa la intersección entre un valor del tiempo y un valor de la variable analizada, para nuestro ejemplo cada punto representa la información de las ventas realizadas en cada uno de los diferentes meses.</p>
<p style='text-align: justify;'>4.	La <font color = green><b>línea<b></font> que une los diferentes puntos, la cuál se utiliza para indicar la tendencia obtenida por la variable analizada durante dos valores continuos del intervalo de tiempo.</p>
<br>

<br>
<p style='text-align: justify;'>En nuestro segundo ejemplo hemos utilizado una gráfica de línea para representar simultáneamente dos variables diferentes, el valor total de las ventas realizadas durante cada mes por un encargado y las ganancias que obtuvo el mismo empleado en el mismo intervalo de tiempo.</p>

```python
import pandas as pd
```

Datos Históricos de BIMBOA.MX

https://finance.yahoo.com/quote/BIMBOA.MX/history?p=BIMBOA.MX

```python
cemex = pd.read_csv('data/linea/CEMEXCPO.MX.csv',index_col=0)
bimbo = pd.read_csv('data/linea/BIMBOA.MX.csv',index_col = 0)
cemex.head()
```

```python
cemex.index = pd.to_datetime(cemex.index)
bimbo.index = pd.to_datetime(bimbo.index)
cemex.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(cemex['Close'])
```

```python
# linestyle or ls [ '-' | '--' | '-.' | ':' | 'steps' | ...]
# marker [ '+' | ',' | '.' | '1' | '2' | '3' | '4' ]
plt.plot(cemex['Close'],'b-',linewidth=1)
```

```python
# alpha : El % de la intensidad del color
# color : b- Lineas Azules, g- Líneas Verdes
# label : para las etiquetas que sacamos luego en el cuadro de las leyendas
plt.plot(cemex['Close']      ,'g-',alpha = 1.0, label='a:1.0')
plt.plot(cemex['Close'] * 2  ,'g-',alpha = 0.5, label='a:0.5')
plt.plot(cemex['Close'] *1/2 ,'g-',alpha = 0.1, label='a:0.1')
plt.legend()
```

```python
plt.plot(cemex['Low'],label='Low')
plt.plot(cemex['High'],label='High')
plt.legend()
```

```python
plt.plot(cemex['Low'],label='LowCemex')
plt.plot(cemex['High'],label='HighCemex')
plt.plot(bimbo['High'],label='HighBimbo')
plt.legend()
```

```python
plt.figure()
plt.subplot(211)
plt.plot(cemex['Low'],label='Low')
plt.plot(cemex['High'],label='High')
plt.legend()
plt.subplot(212)
plt.plot(bimbo['Low'],label='Low')
plt.plot(bimbo['High'],label='High')
plt.legend()
```

```python
plt.figure()
ax1 = plt.subplot(211)
plt.plot(cemex['Low'],label='Low')
plt.plot(cemex['High'],label='High')
plt.title('Cemex')
plt.tight_layout()
plt.legend()
ax2 = plt.subplot(212)
plt.plot(bimbo['Low'],label='Low')
plt.plot(bimbo['High'],label='High')
plt.title('Bimbo')
plt.tight_layout()
plt.legend()
```

```python
plt.figure()
ax1 = plt.subplot(211)
plt.plot(cemex['Low'],'r--',label='Low',alpha=0.3)
plt.plot(cemex['High'],'b--',label='High',alpha=0.3)
plt.plot(cemex['Close'],'g-',label='Close',alpha=1)
plt.title('Cemex')
plt.xlim(pd.to_datetime('2019-05-01'), pd.to_datetime('2019-06-01')) # NO JALA como venia, plique la conversión
plt.ylim(7.5, 9.5)
plt.xticks(rotation=30)
plt.tight_layout()
plt.legend()
ax2 = plt.subplot(212)
plt.plot(bimbo['Low'],'r--',label='Low',alpha=0.3)
plt.plot(bimbo['High'],'b--',label='High',alpha=0.3)
plt.plot(bimbo['Close'],'g-',label='Close',alpha=1)
plt.title('Bimbo')
#plt.xlim('2019-05', '2019-06') # NO JALA
plt.xlim(pd.to_datetime('2019-05-01'), pd.to_datetime('2019-06-01'))
plt.ylim(35, 45)
plt.xticks(rotation=30)
plt.tight_layout()
plt.legend()
plt.savefig("graficadelinea.png",dpi=300)
```
