## Seleccionar renglones
#### loc[] vs iloc[]


```python
df.loc[[0,1,2]] # loc accede en base a su etiqueta indice
```

```python
df.iloc[0:3]  # iloc accede en base a su indice/posición; selección tipo LISTAS/STR
```

## Seleccionar columnas

```python
#df.sex.head()      # Las 2 formas son correctas
df["sex"].head()
```

```python
df[["age","sex","target"]].head()   # Se hace con listas de las columnas
```

```python
df.iloc[:,[0,1,-1]].head() # : - Traeme todas las FILAS; [1,1,-1] y éstas COLUMNAS
```

## Seleción condicional

```python
df[df["sex"] == 1].head()
```

## Ordenar

```python
df.sort_values(by="chol")               # Ordena por una sola columna, recibe un solo parametro, 
                                        # puede ser un valor o una lista, CREO QUE TODO PANDAS
                                        # TRABAJA ASI
df.sort_values(by=["chol","oldpeak"])   # Ordena por varias columnas
```

## Groupby

```python
df.groupby("sex").mean()            # Funciona igual que l anterior, un param, y puede ser valor o una lista
#df.groupby(["sex","age"]).mean()   # OK
```

```python
df.groupby("target").count()
```
