3. Elabora un histograma en el que se observe la distribución de datos de la población por país que vive con menos de \\$1.90 dólares al día. El histograma debe contener los siguientes elementos:

    1. Título

    4. 20 bines

```python
plt.hist(df['< $1.90[1]'], bins=20, label='Paises', edgecolor='black', linewidth=1  ) 
plt.xlabel('% < $1.90 Dólares x Día')
plt.ylabel('# Paises')
plt.ylim(0, 100)
plt.title("Población que vive con menos de $1.90 x día")
plt.legend()
plt.show()
```

4. __¿Cuántos son los países que tienen aproximádamente cerca del 0% y 4% de su población viviendo con menos de \$1.90 dólares al día?__

6. Selecciona de tu dataframe llamada **_df_** solo aquellos países que posean a menos del 10% de su población viviendo con menos de \$1.90 dólares al día y guárdala en una variable llamada __dfmenos10__.

Después, selecciona de tu dataframe aquellos países que tengan a más del 10% de su población viviendo con menos de la cantidad mencionada anteriormente y guárdala en una variable llamada __dfmas10__.

```python
dfmenos10 = df['< $1.90[1]']
dfmenos10 = df1[(df1 <= 10)]
print(dfmenos10.head())
print(dfmenos10.describe())

dfmas10 = df1[(df1 > 10)]
print(dfmas10.head())
print(dfmas10.describe())
```

7. Grafica dos histogramas con tus dataframes llamados **_dfmenos10_** y **_dfmas10_** en la cual tenga las siguientes características:
    1. Título
    2. Etiquetas en los ejes.
    4. 20 bines
    5. Color del histograma: Naranja
    6. Color de los bordes del histograma: __Negro__
    7. Grosor de los bordes: 1
    8. Grid con transparencia de 0.3.
    9. Deben estar colocados uno arriba del otro

```python
fig, ax = plt.subplots(2, 1, sharex=False)
ax[0].hist(dfmenos10, bins=20, label='Paises', edgecolor='black', linewidth=1, color='orange', alpha=0.3) 
ax[1].hist(dfmas10, bins=20, label='Paises', edgecolor='black', linewidth=1, color='orange', alpha=0.3) 

ax[0].set_title('Paises con <= 10%')
ax[1].set_title('Paises con > 10%')

#plt.title("Población que vive con menos de $1.90 x día")
ax[0].legend()
ax[1].legend()

ax[0].set_xlabel('% < $1.90 CON MENOS DEL 10%')
ax[1].set_xlabel('% < $1.90 CON MAS DEL 10%')

ax[0].set_ylabel('# Paises')
ax[1].set_ylabel('# Paises')

fig.tight_layout(pad=1.0)
plt.show()
```

9. Genera una sola gráfica en la que se muestren tres histogramas correspondientes a la distribución de porcentaje de población que gana menos de \\$1.90, \\$3.20 y \\$5.50 dólares respectivamente. Las gráficas deben poseer las siguientes características:

```python
fig, ax = plt.subplots() #2, 1, sharex=False)
ax.hist(df['< $1.90[1]'], label='% < $1.90', linewidth=1, alpha=0.3) 
ax.hist(df['< $3.20[5]'], label='% < $3.20', linewidth=1, alpha=0.3) 
ax.hist(df['< $5.50[6]'], label='% < $5.50', linewidth=1, alpha=0.3) 

plt.title('% de Población en Pobreza')
plt.xlabel('%')
plt.ylabel('# Países')
#fig.tight_layout(pad=1.0)
plt.legend()
plt.savefig('Hist_Gen.png', dpi=100)
plt.show()
```
